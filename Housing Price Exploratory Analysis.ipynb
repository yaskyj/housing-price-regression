```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
pd.options.display.max_columns = 2000
pd.options.display.max_rows = 2000
```

```python
train = pd.read_csv('train.csv')
```

```python
train.head()
```

```python
train.info()
```

```python
train.describe()
```

```python
### The target appears skewed
train['SalePrice'].hist(bins=50)
plt.show()
```

```python
train['target'] = np.log1p(train["SalePrice"])
train['target'].hist(bins=50)
plt.show()
```

```python
### Create overall features without Id and target
features = [c for c in train.columns if c not in['Id','SalePrice', 'target']]
```

```python
### Show histograms for all numerical features
train[features].hist(bins=50, figsize=(30,20))
plt.show()
```

```python
### Create list containing the numerical and categorical columns
cats = []
nums = []
for c, dtype in zip(train[features].columns, train[features].dtypes):
    if dtype == 'object':
        cats.append(c)
    else:
        nums.append(c)
```

```python
from scipy.stats import skew
skewed_feats = train[nums].apply(lambda x: skew(x.dropna())) #compute skewness
skewed_feats = skewed_feats[skewed_feats > 0.75]
skewed_feats = skewed_feats.index
# train[skewed_feats] = np.log1p(train[skewed_feats])
train[skewed_feats].hist(bins=50, figsize=(30,20))
plt.show()
```

```python
### Showing nulls for categorical features
train[cats].isnull().sum().sort_values(ascending=False)
```

```python
### Showing nulls for numerical features
train[nums].isnull().sum().sort_values(ascending=False)
```

```python
### 
from sklearn.preprocessing import Imputer
imputer = Imputer()
imputer = imputer.fit(train[nums])
train[nums] = imputer.transform(train[nums])
```

```python
### Add dummy variables for categories
train = pd.get_dummies(train, columns = cats)
```

```python
### Calculate correlations among all columns
corr_matrix = train.corr()
```

```python
corr_matrix['SalePrice'].sort_values(ascending=False)
```
